```python
def periodo_von_neumann(x):
    contador = 1
    lista_generada = [x]
    for i in range(10000):
        x = (x**2 // 100) % 10000 # n // 10^k elimina los últimos k dígitos del número y n % 10^k da los últimos k dígitos del número.
        if x in lista_generada:
            break
        lista_generada.append(x)
        contador += 1
    
    return contador

print(periodo_von_neumann(3009))
print("------------------------")
print(periodo_von_neumann(7600))
print("------------------------")
print(periodo_von_neumann(1234))
print("------------------------")
print(periodo_von_neumann(4321))
```

```python
import matplotlib.pyplot as plt

def generador_nuevo(a, c):
    """
    Generador de números pseudoaleatorios utilizando un generador lineal congruencial
    Pueden cambiar y x por cualquier semilla que se les ocurra
    """
    y = 1
    x = 1
    z = (y + x) % 32
    ys = [y]
    xs = [x]
    zs = [z]

    for i in range(1000):
        y = (5 * y + c) % 32
        x = (a * x) % 31
        z = (y + x) % 32
        ys.append(y)
        xs.append(x)
        zs.append(z)

    return ys, xs, zs

# Test generador_nuevo
ys, xs, zs = generador_nuevo(3, 1)

fig, ax = plt.subplots(3, 1)
ax[0].plot(ys[:-1], ys[1:], 'ro', label='y')
ax[1].plot(xs[:-1], xs[1:], 'bo', label='x')
ax[2].plot(zs[:-1], zs[1:], 'ko', label='z')
plt.show()
```

```python
def buscar_raices_primitivas():
    """
    Busqueda de raíces primitivas de 31.
    """
    raices_primitivas = []
    for a in range(1, 31):
        if a**15 % 31 == 1:
            continue
        if a**10 % 31 == 1:
            continue
        if a**6 % 31 == 1:
            continue
        raices_primitivas.append(a)
    
    return raices_primitivas

print(buscar_raices_primitivas())
```

```python
def raices_primitivas():
    """
    Búsqueda de raíces primitivas de 71
    70 = 7 * 5 * 2 
    70 / 7 = 10
    70 / 5 = 14
    70 / 2 = 35
    """
    raices_primitivas = []
    for a in range (1, 71):
        if (a**35) % 71 == 1:
            continue
        if (a**14) % 71 == 1:
            continue
        if (a**10) % 71 == 1:
            continue
        raices_primitivas.append(a)
    return raices_primitivas
print(raices_primitivas())
```

```python
import matplotlib.pyplot as plt
import math

def randu(a, M):
    x = 1
    lista = [x]
    for i in range(50000):
        x = (a * x) % M
        lista.append(x)
    
    return lista

a = 2**16+3 #7**5
M = 2**30 - 1 # 2 ** 31 - 1
lista = randu(a, M)

xs = lista[0::3]
ys = lista[1::3]
zs = lista[2::3]

# Plot 3D de los puntos generados con ternas (u_i, u_i+1, u_i+2)
# Rotarlo para encontrar los hiperplanos
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c='r', marker='o')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

# Estimacion de puntos que entran en una esfera de radio M/10
# en el cubo de lado 
# Esto deberia darnos cercano a el cociente volumen de la esfera / volumen del cubo
contador = 0
for i in range(0, int(50001 / 3)):
    if (xs[i] - M / 2)**2 + (ys[i] - M / 2)**2 + (zs[i] - M / 2)**2 <= (M / 10)**2:
        contador += 1

print(f"Estimacion de puntos en la esfera: {contador / (50001 / 3)}")
print(4 * math.pi / 3000)

```

# Ejercicio 2

```python
import random

def juego_azar():
    u = random.random()
    if u < 1/2 :
        w1 = random.random()
        w2 = random.random()
        x = w1 + w2
    else: 
        w1 = random.random()
        w2 = random.random()
        w3 = random.random()
        x = w1 + w2 + w3
    return x

Nsim = 1000000
prob = 0
for i in range (0, Nsim):
    y = juego_azar()
    if y >= 1:
        prob += 1
prob /= Nsim
print(prob)
```

```python
import random

def traga_monedas():
    u = random.random()
    if u < 1/3 :
        w1 = random.random()
        w2 = random.random()
        x = w1 + w2
    else: 
        w1 = random.random()
        w2 = random.random()
        w3 = random.random()
        x = w1 + w2 + w3
    return x

Nsim = 1000000
prob = 0
for i in range (0, Nsim):
    y = traga_monedas()
    if y <= 2:
        prob += 1
prob /= Nsim
print(prob)
```

```python
import numpy as np
import random
import math

# Con esta función simulo la elección de las cajas
def elegir_caja():
    probabilidad = random.uniform(0, 1) # Devuelve un número aleatorio entre 0 y 1
    if probabilidad < 0.4:
        return 1
    elif probabilidad < 0.72:
        return 2
    else:
        return 3

# Cada caja tiene un tiempo de espera que sigue una distribución exponencial
def caja_espera(caja):
    if caja == 1:
        return np.random.exponential(3) 
    elif caja == 2:
        return np.random.exponential(4)
    elif caja == 3:
        return np.random.exponential(5)
    else:
        raise ValueError("Caja no válida. Debe ser 1, 2 o 3.")

# Con esta función simulo la elección de las cajas
def elegir_caja():
    probabilidad = random.uniform(0, 1) # Devuelve un número aleatorio entre 0 y 1
    if probabilidad < 0.4:
        return 1
    elif probabilidad < 0.72:
        return 2
    else:
        return 3


# Acá simulo la elección de las elecciones de cajas y sus esperas
def simular_espera(iteraciones):
    tiempos_espera = []
    cajas = []
    for _ in range(iteraciones):
        caja = elegir_caja()
        tiempo = caja_espera(caja)
        tiempos_espera.append(tiempo)
        cajas.append(caja)
    
    return tiempos_espera, cajas

def calcular_probabilidades(tiempos_espera, cajas):
    total = len(tiempos_espera)
    menos_4 = sum(1 for t in tiempos_espera if t < 4)
    prob_menos_4 = menos_4 / total
    
    cajas_mas_4 = [0, 0, 0]
    for i in range(total):
        if tiempos_espera[i] > 4:
            cajas_mas_4[cajas[i] - 1] += 1
    
    prob_cajas_mas_4 = [cajas_mas_4[i] / total for i in range(3)]
    
    return prob_menos_4, prob_cajas_mas_4

print("Simulando el problema...")
tiempos_espera, cajas = simular_espera(1000) 
prob_menos_4, prob_cajas_mas_4 = calcular_probabilidades(tiempos_espera, cajas)
print(f"Probabilidad de que un cliente espere menos de 4 minutos: {prob_menos_4:.2f}")

# Ahora para calcular la probabilidad de que el cliente haya elegido cada una de las cajas dado que esperó más de 4 minutos

print("Probabilidades de que el cliente haya elegido cada caja dado que esperó más de 4 minutos:")
for i in range(3):
    prob = prob_cajas_mas_4[i] / sum(prob_cajas_mas_4)
    print(f"Caja {i + 1}: {prob:.2f}")
    
```
